```python
with open('data.csv') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    
    language_counter = Counter()
    
    for row in csv_reader:
        language_counter.update(row['LanguagesWorkedWith'].split(';'))
    
# print(language_counter)
languages = []
popularity = []
# print(language_counter.most_common(15))
for item in language_counter.most_common(15):
    languages.append(item[0])
    popularity.append(item[1])
    
# print(languages)
# print(popularity)
plt.bar(languages, popularity)
plt.xlabel('Programming Languages')
plt.ylabel('No of people who use')
plt.title('Most popular lang')
```

```python
languages.reverse()
popularity.reverse()
plt.barh(languages, popularity)
plt.xlabel('No of people who use')
plt.title('Most popular languages')
plt.grid(True)
# plt.tight_layout()
```

```python
data = pd.read_csv('data_lang.csv')
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']
plt.plot(ages, dev_salaries, color= "#fc4f30", linestyle = '--', 
         label = 'All Devs')
plt.plot(ages, py_salaries, color = "#008fd5", label = 'Python')
overall_median = 57287
plt.fill_between(ages, py_salaries, overall_median, where=
                 (py_salaries>overall_median),interpolate = True,
                 alpha = 0.25) #alpha gives transparencu
plt.fill_between(ages, py_salaries, overall_median, where=
                 (py_salaries<overall_median),interpolate = True,
                 color='yellow',alpha = 0.25)
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.title('Median Salary by Age')
plt.legend()
plt.show()
```
